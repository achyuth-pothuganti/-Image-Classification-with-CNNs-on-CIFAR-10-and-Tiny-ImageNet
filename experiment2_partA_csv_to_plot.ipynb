```python
import os

custom_model_base_path_CIFAR100 = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\Custom\\CIFAR100\\'

custom_model_base_path_IMAGENET = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\Custom\\ImageNet\\'

resnet_base_path_CIFAR10 = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\ResNet\\CIFAR10\\'
resnet_base_path_CIFAR100 = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\ResNet\\CIFAR100\\'
resnet_base_path_IMAGENET = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\ResNet\\ImageNet\\'

vgg_base_path_CIFAR10 = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\VGG19\\CIFAR10\\'
vgg_base_path_CIFAR100 = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\VGG19\\CIFAR100\\'
vgg_base_path_IMAGENET = 'C:\\Users\\ashna\\Desktop\\CS 5567\\Final\\Exp2_partA_results\\VGG19\\ImageNet\\'

paths = [custom_model_base_path_CIFAR100, 
         custom_model_base_path_IMAGENET, 
         resnet_base_path_CIFAR10, 
         resnet_base_path_CIFAR100,
         resnet_base_path_IMAGENET,
         vgg_base_path_CIFAR10,
         vgg_base_path_CIFAR100,
         vgg_base_path_IMAGENET]

for path in paths: 
    # Check if the path exists
    if os.path.exists(path):
        print("The path exists.")
    else:
        print("The path does not exist.")
```

```python
import os
import glob
import matplotlib.pyplot as plt
import pandas as pd

for save_plot_path in paths:
    # Define the pattern to match CSV files
    pattern = '*.csv'

    # Construct the full path pattern
    path_pattern = os.path.join(save_plot_path, pattern)

    # Find all CSV files in the directory
    csv_files = glob.glob(path_pattern)

    # Check if any CSV files are found
    if csv_files:
        # Save the name of the first CSV file found
        first_csv_file = os.path.basename(csv_files[0])
        print(f"Found CSV file: {first_csv_file}")
    else:
        print("No CSV files found.")
        first_csv_file = None  # Assign None if no file is found
    full_path = os.path.join(save_plot_path + first_csv_file)
    df = pd.read_csv(full_path)
    # plotting
    plt.figure(figsize=(14, 7))
    # Subplot 1: Training and Validation Accuracy
    plt.subplot(1, 2, 1)
    plt.plot(df['accuracy'], label='Training Accuracy', marker='o')
    plt.plot(df['val_accuracy'], label='Validation Accuracy', marker='o')
    plt.title('Training and Validation Accuracy')
    plt.xlabel('Epochs')
    plt.ylabel('Accuracy')
    plt.legend()
    # Subplot 2: Training and Validation Loss
    plt.subplot(1, 2, 2)
    plt.plot(df['loss'], label='Training Loss', marker='x')
    plt.plot(df['val_loss'], label='Validation Loss', marker='x')
    plt.title('Training and Validation Loss')
    plt.xlabel('Epochs')
    plt.ylabel('Loss')
    plt.legend()
    plt.tight_layout()
    # Save the plot to a file
    filename = 'plot.png'  # You can change the file name and format (e.g., .png, .jpeg, .svg, etc.)
    save_path = os.path.join(save_plot_path, filename)
    plt.savefig(save_path, format='png', dpi=300)
    

```
